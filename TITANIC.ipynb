# Titanic Dataset - Exploratory Data Analysis
**Objective:** Extract insights about survival rates, passenger demographics, and relationships between variables.  
**Tools Used:** Python (Pandas, NumPy, Matplotlib, Seaborn)


## Importing Libraries and Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

df = pd.read_csv("titanic.csv")
df.head()
```

## Dataset Overview :

```python
print(df.shape)
df.info()
df.describe()
df.isnull().sum()
```

## Categorical Variable Counts

```python
df['Survived'].value_counts()
df['Pclass'].value_counts()
df['Embarked'].value_counts()
```

## Univariate Analysis

```python
sns.histplot(df['Age'].dropna(),kde=True,bins=30, color = 'red')
plt.title("Age Distribution")
plt.show()
```

## Bivariate Analysis

```python
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set2')
plt.title("Survial by Gender")
plt.show()
```

## Correlation & Relationships

```python
plt.Figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

- Women had a ~75% survival rate, much higher than men (~19%).
- Younger passengers (<15) had a higher survival rate.
- First-class passengers were more likely to survive.

## Summary of Findings
1. Gender is a strong predictor of survival.
2. Higher-class passengers had higher survival rates.
3. Age distribution is skewed toward younger adults.
4. Missing values are mainly in `Age`, `Cabin`, and `Embarked`.
